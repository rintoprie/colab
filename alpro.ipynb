```python
TB1 = int(input("Masukkan nilai TB1: "))
TB2 = int(input("Masukkan nilai TB2: "))
UAS = int(input("Masukkan nilai UAS: "))
nilai = TB1 * 0.3 + TB2 * 0.3 + UAS * 0.4
if nilai > 80:
  grade = 'A'
elif nilai > 70:
  grade = 'B'
elif nilai > 60:
  grade = 'C'
else:
  grade = 'D'
print("Grade: " + grade)
```

```python
def is_prima (x):
  for i in range(2, x):
    if x % i == 0:
      return False

  return True

for i in range(2,10):
  if is_prima(i):
    print(i)
```

```python
for n in range(2,10):
  bilangan = n
  prima = True
  for i in range(2,bilangan):
    if (bilangan % i == 0):
      prima = False
  if (prima):
    print(bilangan)

```

```python
bilangan = 2
count = 1
n = 10
while count <= n:
  prima = True
  for i in range(2,bilangan):
    if (bilangan % i == 0):
      prima = False
  if (prima):
    print(bilangan)
    count += 1
  bilangan += 1
```
